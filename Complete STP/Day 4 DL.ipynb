```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.linspace(-100,100,500)
```

```python
plt.plot(y)
plt.show()
```

```python
y_sigmoid = 1 / (1 + np.e**(-1*y))
```

```python
plt.plot(y_sigmoid)
plt.title('Sigmoid Activation')
plt.annotate(text = '''1 / (1 + np.e**(-1*y))''',xy = (50,0.5))
plt.show()
```

```python
y_sigmoid.min()
```

```python
y_sigmoid.max()
```

# Tanh Activation

# 

```python
y = np.linspace(-5,5,500)
tanh_y = (np.e**(y) - np.e**(-1*y))/(np.e**(y) + np.e**(-1*y))
```

```python
plt.plot(tanh_y)
plt.title('Tanh Activation')
plt.show()
```

```python
tanh_y.min()
```

```python
tanh_y.max()
```

```python
#sigmoid activation func: output layer: Binary result>> 0,1: hidden layer never use
```

```python
# relu
```

```python
# np.where(y<0,0,y)
```

```python
final_y = []
for i in y:
    if i<0:
        final_y.append(max(0,i))
    else:
        final_y.append(i)
print(final_y)
```

```python
print(np.where(y<0,0,y))
```

```python
plt.plot(final_y)
plt.title('Relu Activation')
plt.show()
```

```python
plt.plot(np.where(y<0,0,y))
plt.title('Relu Activation')
plt.show()
```

```python
# parametric relu
```

```python
plt.plot(np.where(y<0,0.01*y,y))
plt.title('Parametric Relu Activation')
plt.show()
```

```python
y_final = np.where(y<0,0.01*y,y)
```

```python
y_final.min()
```

```python
y_final.max()
```

# 

```python
elu_y  = 0.01*(np.e**y - 1)
```

```python
plt.plot(elu_y)
plt.title('Elu Activation')
plt.show()
```

```python
plt.plot(np.where(y<0,0.1*y,y))
plt.title('Leaky Relu Activation')
plt.axhline(0,color = 'r')
plt.show()
```
