# Part 1 Solving Problem Using ML

# Important Modules Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
lbc = load_breast_cancer()
```

```python
X = lbc.data
```

```python
y = lbc.target
```

```python
# MinMaxScaler apply to X features not target y
```

```python
scaler = MinMaxScaler()
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
scaled_X.min()
```

```python
scaled_X.max()
```

```python
# train test split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_X,y,test_size=0.2,random_state=42)
```

# Step ML model creation

```python
from sklearn.linear_model import LogisticRegression
```

```python
ml_model = LogisticRegression()
```

```python
ml_model.fit(X_train,y_train)
```

```python
y_pred = ml_model.predict(X_test)
```

```python
ml_model.score(X_test,y_test)
```

```python
ml_model.predict_proba(X_test)[0]
```

# Part 2 Solving problem Using Deep Leaning

```python
scaled_X.shape[1]*2
```

```python
scaled_X.shape
```

```python
# Module Import
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import warnings
warnings.filterwarnings('ignore')
```

```python
X_train[0].shape
```

```python
# Create Model

model = Sequential([
    Dense(40, input_shape = X_train[0].shape, activation='relu'), # input layer + 1st hidden layer
    Dense(100, activation='relu'),# 2nd Hidden layer
    Dense(50, activation='relu'), #3rd Hidden Layer
    Dense(1, activation='sigmoid')# Output layer: 0/1: Sigmoid
])
print('Model Created Successfully!!')
```

```python
# Model Compile
model.compile(optimizer=Adam(learning_rate=0.001),loss = 'binary_crossentropy', metrics=['accuracy'])
print('Model compiled')
```

```python
# Train your Model
model.fit(X_train,y_train,epochs=50, batch_size=50, validation_split=0.2)
```

```python
plt.title('Accuracy vs Epochs')
plt.plot(model.history.history['accuracy'], label = 'Training accuracy')
plt.plot(model.history.history['val_accuracy'], label = 'Validation accuracy')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.show()
```

```python
plt.title('Loss vs Epochs')
plt.plot(model.history.history['loss'], label = 'Training loss')
plt.plot(model.history.history['val_loss'], label = 'Validation loss')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

```python
plt.title(' Training Accuracy vs Training Loss')
plt.plot(model.history.history['accuracy'], label = 'Training accuracy')
plt.plot(model.history.history['loss'], label = 'Training loss')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.show()
```

```python
plt.title(' Val Accuracy vs Val Loss')
plt.plot(model.history.history['val_accuracy'], label = 'Val accuracy')
plt.plot(model.history.history['val_loss'], label = 'Val loss')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.show()
```
